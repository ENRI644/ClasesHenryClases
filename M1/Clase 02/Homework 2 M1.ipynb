1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
```

```python
dicc_emisiones = {}
cod_pais = []
nom_pais = []
region = []
anio = []
co2 = []
co2_percapita = []
```

```python
encabezado = archivo.readline()
for i in archivo:
    i=i.rstrip() #elimina los espacios en blanco y los saltos de linea (/n)
    aux=i.split("|") #separo el renglon, por el separador |
    encabezado = archivo.readline()
    if encabezado == True:
        continue
    else: 
        anio.append(int(aux[3]))
    
    if encabezado == True:
        continue
    elif aux[4] == "" or aux[4] == None:
        co2.append(0.0)
    else:    
        co2.append(float(aux[4].replace(".","").replace(",",".")))
    
    if encabezado == True:
        continue
    elif aux[5] == "" or aux[5] == None or aux[5] == " ":
        co2.append(0.0)
    else:
        co2_percapita.append(float(aux[5].replace(".","").replace(",",".")))
    
    cod_pais.append(str(aux[0]))
    nom_pais.append(str(aux[1]))
    region.append(str(aux[2]))

dicc_emisiones = {'cod_pais' : [cod_pais],
    'nom_pais' : [nom_pais],
    'region' : [region],
    'anio' : [(anio)],
    'co2' : [(co2)],
    'co2_percapita' : [co2_percapita]}

print(dicc_emisiones)

archivo.close()
```
